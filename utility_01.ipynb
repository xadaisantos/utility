```python
def fat(n):
    """
    Função fatorial usando recursão
    """
    if n == 0 or n == 1: # Essa parte da lógica serve para parar a recursão
        return 1
    return n * fat(n - 1)

def exp(a, x):
    """
    Função exponencial usando recursão
    """
    if x == 0:
        return 1
    elif x == 1:
        return a
    else:
        return a * exp(a, x - 1)
    
def linearsearch(array, valor):
    """
    Busca linear
    """
    for x in range(len(array)):
        if array[x] == valor:
            return array.index(valor)
    else:
        return -1
    
def fibfib(n):
    """
    Returns the Fibonacci sequence up to n
    """
    
    lista = [0, 1]
    
    if n == 1:
        return 0
    elif n == 2:
        return 1
    elif n > 2:
        for num in range(0, n - 2):
            lastElement = lista[-1] + lista[-2]
            lista.append(lastElement)
        return lista
    else:
        return 'not possible'
    
def divisores(teste):
    if teste == 0:
        return 'all'
    if teste == 1:
        return 1
    divisores = []
    for num in range(2, teste + 1):
        if teste % num == 0:
            divisores.append(num)
    return divisores

def primo(teste):    
    divisores = set()
    if teste == 0 or teste == 1:
        return False
    if teste == 2:
        return True
    for num in range(2, (teste // 2) + 1):
        if teste % num == 0:
            divisores.add(False)
        else:
            divisores.add(True)
            
    return all(divisores)

def rot13(string):
    """
    Converts the string using the encryption code ROT13.
    ROT13 swaps the letter of the first half of the alphabet with the second half.
    """
    alfa = 'abcdefghijklmnopqrstuvwxyz'
    middle = len(alfa) // 2
    alfa1 = alfa[:middle]
    alfa2 = alfa[middle:]
    result = ''
    
    for letter in string.lower():
        if letter in alfa1:            
            index = alfa1.index(letter)
            new_letter = alfa2[index]
            result += new_letter
        elif letter in alfa2:
            index = alfa2.index(letter)
            new_letter = alfa1[index]
            result += new_letter
        else:
            result += letter

    return result
```
